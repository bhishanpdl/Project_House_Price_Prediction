# Load the libraries

```python
import numpy as np
import pandas as pd
import glob
import os
import sys

# plotting
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(color_codes=True)
sns.set()
sns.set_style('darkgrid')
%config InlineBackend.figure_format = 'svg'
```

# Load the data

```python
files = glob.glob('outputs/*.csv')
df = pd.concat([pd.read_csv(f) for f in files])
df['name'] = files

regex = r"outputs/(?P<model>\w+)_scaling_(?P<scaling>\w+)_logsq_(?P<logsq>\d)_logtarget_(?P<logtarget>\d).csv"

df1 = df['name'].str.extract(regex,expand=True)
df = df.drop('name',axis=1)
df = pd.concat([df1,df],axis=1)
df = df.sort_values('Adjusted R-squared',ascending=False)
df = df.reset_index(drop=True)
df
```
